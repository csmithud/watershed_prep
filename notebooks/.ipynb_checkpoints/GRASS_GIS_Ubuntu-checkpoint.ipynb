```python
Stream_Threshold_NonLocal = 600000
Stream_Threshold_Local= 2000
Stream_Thresholds = [Stream_Threshold_Local, Stream_Threshold_NonLocal]
```

```python
#Variables
Return_Interval =100.0
#Accumulation Threshold for Basins
Threshold = 60000

#Vector and Rater names
file_heatmap_RI = '%s_%s_heatmap_RI' %(Project_Area, Model_Type)
file_shade = select_data+'_%sm_shade' % DEM_Res
file_bounding_box = 'v_'+select_data+'_%sm_bounding_box' % DEM_Res
file_NHDPLUS_bounding_box = 'v_'+select_data+'_NHDPLUS'
file_burned_DEM = select_data+'_%sm_burned_DEM'% DEM_Res
file_burned_shade = select_data+'_%sm_burned_shade'% DEM_Res

file_burned_accum = select_data+'_%sm_burned_accum'% DEM_Res
file_burned_draindir = select_data+'_%sm_burned_draindir'% DEM_Res
file_burned_basin = select_data+'_%sm_burned_basin'% DEM_Res
file_burned_streams_der = select_data+'_%sm_burned_streams_der'% DEM_Res
file_burned_streams_der_t = select_data+'_%sm_burned_streams_der_t'% DEM_Res
file_burned_streams_der_t_v = 'v_'+select_data+'_%sm_burned_streams_der_t'% DEM_Res

file_distance_stream = select_data+'_%sm_distance_stream'% DEM_Res
file_elevation_stream = select_data+'_%sm_elevation_stream'% DEM_Res
file_final_elevation_stream = select_data+'_%sm_final_elevation_stream'% DEM_Res
file_slope= select_data+'_%sm_slope'% DEM_Res
file_curvature= select_data+'_%sm_curvature'% DEM_Res
file_slope_nonzero = select_data+'_%sm_slope_nonzero'% DEM_Res
file_TWI= select_data+'_%sm_TWI'% DEM_Res
file_contributing_area = select_data+'_%sm_contributing_area'% DEM_Res
file_contributing_area_final = select_data+'_%sm_final_contributing_area'% DEM_Res
file_geo_flood_index = select_data+'_%sm_Geo_FloodIndex'% DEM_Res
```

```python
file_DEM = Project_Area+'_DEM_'+('%sm' % DEM_Res)
file_HeatMap = Project_Area+'_'+ Model_Type+'_heatmap'
Mask = "%s_Mask_Area" % select_data

raster_files = [file_DEM, file_HeatMap, Mask]
vector_files = ['HUC12_Boundaries','v_'+select_data+'_HUC_Boundary', 'NHDPLUS_'+select_data[:4]+'_HU4_GDB']

GRASS_raster_files= [file for line in gs.list_strings(type='raster') for file in [line.split("@")[0]]]
GRASS_vector_files = [file for line in gs.list_strings(type='vector') for file in [line.split("@")[0]]]

for file in raster_files:
    if file not in GRASS_raster_files:
        print('Missing: '+file)
        
for file in vector_files:
    if file not in GRASS_vector_files:
        print('Missing: '+file)

#Import HUC-12 Boundaries
if vector_files[0] not in GRASS_vector_files:
    gs.run_command('v.import', input= vector_dir/'HUC12.shp',  output= 'HUC12_Boundaries')   
#Extract HUC-12 Boundary of Interest   
if vector_files[1] not in GRASS_vector_files:
    gs.run_command('v.extract', input ='HUC12_Boundaries', where= "HUC12='%s'" % select_data, output = 'v_'+select_data+'_HUC_Boundary')  
        

if vector_files[2] not in GRASS_vector_files:
    gs.run_command('v.import', input=  vector_dir/('NHDPLUS_H_'+select_data[:4]+'_HU4_GDB.gdb'), layer ='NHDFlowline',  output= 'NHDPLUS_'+select_data[:4]+'_HU4_GDB')
    #gs.run_command('v.clip', input='NHDPLUS_'+select_data[:4]+'_HU4_GDB', clip='v_'+select_data+'_HUC_Boundary', output= 'v_'+select_data+'_NHDPLUS')

    
if raster_files[0] not in GRASS_raster_files:
    gs.run_command('r.in.gdal', input= outputs_dir/DEM_Project_Area, output= Project_Area+'_DEM_'+('%sm' % DEM_Res))
    
HeatMap_Clipped_Reprojected = os.path.join(raster_dir, 'HeatMap','{0}_{1}_heatmap_clipped.tif'.format(Project_Area, Model_Type))
if raster_files[1] not in GRASS_raster_files:
    gs.run_command('r.in.gdal', input=HeatMap_Clipped_Reprojected,  output= Project_Area+'_'+ Model_Type+'_heatmap')
    
if raster_files[2] not in GRASS_raster_files:
    gs.run_command('r.mask',vector='v_'+select_data+'_HUC_Boundary')
    gs.mapcalc( "%s_Mask_Area = -%s/%s*100000" % (select_data, file_DEM,file_DEM))
    gs.run_command('r.mask',flags ='r')
    
#Use for listing attributes in a shapefile
#gs.parse_command('v.info', map='Boundary_HUC12', flags='c')

#To Rename Files
#gs.run_command('g.rename',raster="DesPlaines_Pluvial_heatmap,Des_Plaines_Pluvial_heatmap") 

#For deleting and removing raster and vector files
#gs.run_command('g.remove', type ='vector', name = 'Boundary071200040505', flags ='f')

#Retrieve Layer List 
print(fiona.listlayers(vector_dir/('NHDPLUS_H_'+select_data[:4]+'_HU4_GDB.gdb')))
```

```python
region = gs.parse_command('g.region',vector = 'v_'+select_data+'_HUC_Boundary', res = 3, flags='pg')
Buffer = round(min([int(region['cols']),int(region['rows'])])*0.075)
gs.parse_command('g.region',vector = 'v_'+select_data+'_HUC_Boundary', res = 3, flags='pg', grow=Buffer)
gs.parse_command('g.proj', flags='p')
```

```python
#Create Labels from Heatmap
if file_heatmap_RI not in GRASS_raster_files:
    gs.mapcalc("%s = if(%s_%s_heatmap >1/%s, 1, 0)" %(file_heatmap_RI, Project_Area, Model_Type,Return_Interval) )
    print('Created: '+file_heatmap_RI)
    
#Create Shaded Relief Map
if file_shade not in GRASS_raster_files:
    gs.run_command('r.relief', input=Project_Area+'_DEM_'+('%sm' % DEM_Res), output=file_shade, zscale=1.5)
    print('Created: '+file_shade)

if file_NHDPLUS_bounding_box not in GRASS_raster_files:
    gs.run_command('v.in.region', output = file_bounding_box)
    gs.run_command('v.clip', input= 'NHDPLUS_'+select_data[:4]+'_HU4_GDB', clip= file_bounding_box, output= file_NHDPLUS_bounding_box)
    
#Burn Known Streams into the DEM
if file_burned_DEM not in GRASS_raster_files:
    gs.run_command('r.carve', raster= Project_Area+'_DEM_'+('%sm' % DEM_Res), vector=file_NHDPLUS_bounding_box, width="3", depth="10.0", output= file_burned_DEM)
    print('Created: '+file_burned_DEM)
    
#Created Shaded Relief Map of the Burned DEM
if file_burned_shade not in GRASS_raster_files:
    gs.run_command('r.relief', input=file_burned_DEM, output=file_burned_shade, zscale=1.5)
    print('Created: '+file_burned_shade)    

#Calculate Watershed Properties for the Burned DEM
#--- Accumulation, Drainage Direction, and Basins
if file_burned_accum not in GRASS_raster_files:
    gs.run_command('r.watershed', elevation=file_burned_DEM, threshold=Threshold, accumulation= file_burned_accum,
                   drainage = file_burned_draindir, basin=file_burned_basin, flags='a')
    print('Created: '+file_burned_accum)

#Calulate the topographic wetness Index

if file_slope  not in GRASS_raster_files:
    gs.run_command('r.slope.aspect', elevation = Project_Area+'_DEM_'+('%sm' % DEM_Res), slope = file_slope)
    gs.mapcalc("%s = %s + 0.0001" %(file_slope_nonzero,file_slope))
    gs.mapcalc('%s = log(%s/tan(%s), 2.718282)' % (file_TWI, file_burned_accum, file_slope_nonzero))
    print('Created: '+file_TWI)

#Calculate curvature
if file_curvature  not in GRASS_raster_files:
    gs.run_command('r.slope.aspect', elevation = Project_Area+'_DEM_'+('%sm' % DEM_Res), slope = file_slope, pcurvature =file_curvature )
    
GRASS_raster_files= [file for line in gs.list_strings(type='raster') for file in [line.split("@")[0]]]
GRASS_vector_files = [file for line in gs.list_strings(type='vector') for file in [line.split("@")[0]]]
```

```python
mask_array = garray.array(Mask)
accum_array = garray.array(file_burned_accum)
accum_masked_array  = ma.masked_where(mask_array >-100000, accum_array)
#accum_colored_data = colorize(accum_masked_array, cmap='terrain')
accum_array[accum_masked_array.mask]=np.NaN
```

```python
workers = multi.cpu_count()

mapcalc_list = []
rthin_list = []
rtovect_list = []
rstreamdistance_list = []
mapcalc = Module("r.mapcalc", overwrite=True, run_=False)
rthin =  Module("r.thin", overwrite=True, run_=False)
rtovect = Module("r.to.vect", overwrite=True, run_=False)
rstreamdistance = Module("r.stream.distance", overwrite=True, run_=False)



#Calculate Stream Network Raster
queue = ParallelModuleQueue(nprocs=workers)
for i, task in enumerate(jobs):
    new_mapcalc = copy.deepcopy(mapcalc)
    mapcalc_list.append(new_mapcalc)
    m = new_mapcalc(expression = "%s = if(abs(%s) > %s, 1, null())" 
                                  %(file_burned_streams_der+'_'+task,
                                    file_burned_accum, Stream_Thresholds[i]))
    queue.put(m)
    print(queue.get_num_run_procs())
queue.wait()
mapcalc_list = queue.get_finished_modules()
print(mapcalc_list)
for mapcalc in mapcalc_list:
     print(mapcalc.popen.returncode)

#Thin the stream raster 
queue = ParallelModuleQueue(nprocs=workers)
for i, task in enumerate(jobs):
    new_rthin = copy.deepcopy(rthin)
    rthin_list.append(new_rthin)
    m = new_rthin(input = file_burned_streams_der+'_'+task,
                  output = file_burned_streams_der_t+'_'+task )
    queue.put(m)
    print(queue.get_num_run_procs())
queue.wait()
rthin_list = queue.get_finished_modules()
print(rthin_list)
for rthin in rthin_list:
     print(rthin.popen.returncode)
        
#Convert raster to vector  
queue = ParallelModuleQueue(nprocs=workers)
for i, task in enumerate(jobs):
    new_rtovect = copy.deepcopy(rtovect)
    rtovect_list.append(new_rtovect)
    m = new_rtovect(input =file_burned_streams_der_t+'_'+task,
                    output = file_burned_streams_der_t_v+'_'+task,
                    type="line", flags='s')
    queue.put(m)
    print(queue.get_num_run_procs())
queue.wait()
rtovect_list = queue.get_finished_modules()
print(rtovect_list)
for rtovect in rtovect_list:
     print(rtovect.popen.returncode)

#Calculate Height and Distrance from both the Local and NonLocal Drainage AND
#Calculate the contributing area for both the local and nonocal drainage
queue = ParallelModuleQueue(nprocs=workers)
for i, task in enumerate(jobs):
    new_rstreamdistance = copy.deepcopy(rstreamdistance)
    rstreamdistance_list.append(new_rstreamdistance)
    m = new_rstreamdistance(direction=file_burned_draindir,
                        stream_rast=file_burned_streams_der_t+'_'+task,
                        elevation=  Project_Area+'_DEM_'+('%sm' % DEM_Res),
                        distance = file_distance_stream+'_'+task,
                        difference = file_elevation_stream+'_'+task)
    queue.put(m)
    print(queue.get_num_run_procs())
    
    new_rstreamdistance = copy.deepcopy(rstreamdistance)
    rstreamdistance_list.append(new_rstreamdistance)
    m = new_rstreamdistance( direction=file_burned_draindir,
                        stream_rast=file_burned_streams_der_t+'_'+task,
                        elevation=  file_burned_accum,
                        difference = file_contributing_area+'_'+task)
    queue.put(m)
    print(queue.get_num_run_procs())
queue.wait()
rstreamdistance_list = queue.get_finished_modules()
print(rstreamdistance_list)
for rstreamdistance in rstreamdistance_list:
     print(rstreamdistance.popen.returncode)
        
GRASS_raster_files= [file for line in gs.list_strings(type='raster') for file in [line.split("@")[0]]]
GRASS_vector_files = [file for line in gs.list_strings(type='vector') for file in [line.split("@")[0]]]
```
