# Feature Calculation and Training Label Creation

__Description__: 
Features for predicting flood extents are calculated from the following information:
- Digitital Elevation Model (DEM) from the model used to create the training data.
- Stream and river flow vectors, which are flowlines from the National Hydrology Dataset (NHD)

The DEM data is processed with GRASS GIS https://grass.osgeo.org/

This script, creates the GRASS GIS databse, initializes GRASS GIS, intalls any needed modules, and the processes the DEM data. 
With GRASS GIS, the DEM is conditioned by
- Running a Median filter over the topography to remove roads and other man-made obstructions
- Using known stream/river vectors to burn a channel into the DEM

After the DEM is conditioned, GRASS GIS calls the following modules
- r.watersheds
    - Calculates drainage directions and accumulation
- r.watersheds 


__Input__: 
- Parameters: curve number and initial abstraction ratio; the probable maximum precipitation; return intervals for stratified sampling; maximum return interval limit for the stratified sampling bins;  the volume, region, and duration ([See map for volume and region](https://hdsc.nws.noaa.gov/hdsc/pfds/pfds_temporal.html)).
- `Data` folder which contains the following:
    - The Heatmaps
    - The Digital Elevation Models
    - Flow and river lines (in geodatabase format)    

__Output__: 1 Flood prediction indices:
    1) Geomorphic Flood

## GRASS GIS Calculations

```python
#Create Labels from Heatmap
if file_heatmap_RI not in GRASS_raster_files:
    gs.mapcalc("%s = if(%s_%s_heatmap >1/%s, 1, 0)" %(file_heatmap_RI, Project_Area, Model_Type,Return_Interval) )
    print('Created: '+file_heatmap_RI)
    
#Create Shaded Relief Map
if file_shade not in GRASS_raster_files:
    gs.run_command('r.relief', input=Project_Area+'_DEM_'+('%sm' % DEM_Res), output=file_shade, zscale=1.5)
    print('Created: '+file_shade)

if file_NHDPLUS_bounding_box not in GRASS_raster_files:
    gs.run_command('v.in.region', output = file_bounding_box)
    gs.run_command('v.clip', input= 'NHDPLUS_'+select_data[:4]+'_HU4_GDB', clip= file_bounding_box, output= file_NHDPLUS_bounding_box)
    
#Burn Known Streams into the DEM
if file_burned_DEM not in GRASS_raster_files:
    gs.run_command('r.carve', raster= Project_Area+'_DEM_'+('%sm' % DEM_Res), vector=file_NHDPLUS_bounding_box, width="3", depth="10.0", output= file_burned_DEM)
    print('Created: '+file_burned_DEM)
    
#Created Shaded Relief Map of the Burned DEM
if file_burned_shade not in GRASS_raster_files:
    gs.run_command('r.relief', input=file_burned_DEM, output=file_burned_shade, zscale=1.5)
    print('Created: '+file_burned_shade)    

#Calculate Watershed Properties for the Burned DEM
#--- Accumulation, Drainage Direction, and Basins
if file_burned_accum not in GRASS_raster_files:
    gs.run_command('r.watershed', elevation=file_burned_DEM, threshold=Threshold, accumulation= file_burned_accum,
                   drainage = file_burned_draindir, basin=file_burned_basin, flags='a')
    print('Created: '+file_burned_accum)

#Calulate the topographic wetness Index

if file_slope  not in GRASS_raster_files:
    gs.run_command('r.slope.aspect', elevation = Project_Area+'_DEM_'+('%sm' % DEM_Res), slope = file_slope)
    gs.mapcalc("%s = %s + 0.0001" %(file_slope_nonzero,file_slope))
    gs.mapcalc('%s = log(%s/tan(%s), 2.718282)' % (file_TWI, file_burned_accum, file_slope_nonzero))
    print('Created: '+file_TWI)

#Calculate curvature
if file_curvature  not in GRASS_raster_files:
    gs.run_command('r.slope.aspect', elevation = Project_Area+'_DEM_'+('%sm' % DEM_Res), slope = file_slope, pcurvature =file_curvature )
    
GRASS_raster_files= [file for line in gs.list_strings(type='raster') for file in [line.split("@")[0]]]
GRASS_vector_files = [file for line in gs.list_strings(type='vector') for file in [line.split("@")[0]]]
```

```python
workers = multi.cpu_count()

mapcalc_list = []
rthin_list = []
rtovect_list = []
rstreamdistance_list = []
mapcalc = Module("r.mapcalc", overwrite=True, run_=False)
rthin =  Module("r.thin", overwrite=True, run_=False)
rtovect = Module("r.to.vect", overwrite=True, run_=False)
rstreamdistance = Module("r.stream.distance", overwrite=True, run_=False)



#Calculate Stream Network Raster
queue = ParallelModuleQueue(nprocs=workers)
for i, task in enumerate(jobs):
    new_mapcalc = copy.deepcopy(mapcalc)
    mapcalc_list.append(new_mapcalc)
    m = new_mapcalc(expression = "%s = if(abs(%s) > %s, 1, null())" 
                                  %(file_burned_streams_der+'_'+task,
                                    file_burned_accum, Stream_Thresholds[i]))
    queue.put(m)
    print(queue.get_num_run_procs())
queue.wait()
mapcalc_list = queue.get_finished_modules()
print(mapcalc_list)
for mapcalc in mapcalc_list:
     print(mapcalc.popen.returncode)

#Thin the stream raster 
queue = ParallelModuleQueue(nprocs=workers)
for i, task in enumerate(jobs):
    new_rthin = copy.deepcopy(rthin)
    rthin_list.append(new_rthin)
    m = new_rthin(input = file_burned_streams_der+'_'+task,
                  output = file_burned_streams_der_t+'_'+task )
    queue.put(m)
    print(queue.get_num_run_procs())
queue.wait()
rthin_list = queue.get_finished_modules()
print(rthin_list)
for rthin in rthin_list:
     print(rthin.popen.returncode)
        
#Convert raster to vector  
queue = ParallelModuleQueue(nprocs=workers)
for i, task in enumerate(jobs):
    new_rtovect = copy.deepcopy(rtovect)
    rtovect_list.append(new_rtovect)
    m = new_rtovect(input =file_burned_streams_der_t+'_'+task,
                    output = file_burned_streams_der_t_v+'_'+task,
                    type="line", flags='s')
    queue.put(m)
    print(queue.get_num_run_procs())
queue.wait()
rtovect_list = queue.get_finished_modules()
print(rtovect_list)
for rtovect in rtovect_list:
     print(rtovect.popen.returncode)

#Calculate Height and Distrance from both the Local and NonLocal Drainage AND
#Calculate the contributing area for both the local and nonocal drainage
queue = ParallelModuleQueue(nprocs=workers)
for i, task in enumerate(jobs):
    new_rstreamdistance = copy.deepcopy(rstreamdistance)
    rstreamdistance_list.append(new_rstreamdistance)
    m = new_rstreamdistance(direction=file_burned_draindir,
                        stream_rast=file_burned_streams_der_t+'_'+task,
                        elevation=  Project_Area+'_DEM_'+('%sm' % DEM_Res),
                        distance = file_distance_stream+'_'+task,
                        difference = file_elevation_stream+'_'+task)
    queue.put(m)
    print(queue.get_num_run_procs())
    
    new_rstreamdistance = copy.deepcopy(rstreamdistance)
    rstreamdistance_list.append(new_rstreamdistance)
    m = new_rstreamdistance( direction=file_burned_draindir,
                        stream_rast=file_burned_streams_der_t+'_'+task,
                        elevation=  file_burned_accum,
                        difference = file_contributing_area+'_'+task)
    queue.put(m)
    print(queue.get_num_run_procs())
queue.wait()
rstreamdistance_list = queue.get_finished_modules()
print(rstreamdistance_list)
for rstreamdistance in rstreamdistance_list:
     print(rstreamdistance.popen.returncode)
        
GRASS_raster_files= [file for line in gs.list_strings(type='raster') for file in [line.split("@")[0]]]
GRASS_vector_files = [file for line in gs.list_strings(type='vector') for file in [line.split("@")[0]]]
```

```python
#Calculate the actual contributing area value    

mapcalc_list = []
mapcalc = Module("r.mapcalc", overwrite=True, run_=False)

queue = ParallelModuleQueue(nprocs=workers)
for i, task in enumerate(jobs):
    new_mapcalc = copy.deepcopy(mapcalc)
    mapcalc_list.append(new_mapcalc)
    m = new_mapcalc(expression = "%s = if((-%s+%s)<%s,%s,-%s+%s)"
                    %(file_contributing_area_final+'_'+task,
                      file_contributing_area+'_'+task,
                      file_burned_accum,
                      Stream_Thresholds[i],
                      Stream_Thresholds[i],
                      file_contributing_area+'_'+task,
                      file_burned_accum ))
    queue.put(m)
    print(queue.get_num_run_procs())
queue.wait()
mapcalc_list = queue.get_finished_modules()
print(mapcalc_list)
for mapcalc in mapcalc_list:
     print(mapcalc.popen.returncode)
    
## Calculate the Final Stream Elevation   

mapcalc_list = []
mapcalc = Module("r.mapcalc", overwrite=True, run_=False)

queue = ParallelModuleQueue(nprocs=workers)
for i, task in enumerate(jobs):
    new_mapcalc = copy.deepcopy(mapcalc)
    mapcalc_list.append(new_mapcalc)
    m = new_mapcalc(expression = "%s = %s*if(%s<0, 0,1)+.01"
                               % (file_final_elevation_stream+'_'+task,  
                                  file_elevation_stream+'_'+task,
                                  file_elevation_stream+'_'+task))
    queue.put(m)
    print(queue.get_num_run_procs())
queue.wait()
mapcalc_list = queue.get_finished_modules()
print(mapcalc_list)
for mapcalc in mapcalc_list:
     print(mapcalc.popen.returncode)
        
#Calculate the Geomorpholocial flood index

mapcalc_list = []
mapcalc = Module("r.mapcalc", overwrite=True, run_=False)

queue = ParallelModuleQueue(nprocs=workers)
for i, task in enumerate(jobs):
    new_mapcalc = copy.deepcopy(mapcalc)
    mapcalc_list.append(new_mapcalc)
    m = new_mapcalc(expression ="%s = log(%s/(%s*%s), 2.718282)"
                    % (file_geo_flood_index+'_'+task,
                       file_contributing_area_final+'_'+task,
                       file_distance_stream+'_'+task, 
                       file_final_elevation_stream+'_'+task ))
    queue.put(m)
    print(queue.get_num_run_procs())
queue.wait()
mapcalc_list = queue.get_finished_modules()
print(mapcalc_list)
for mapcalc in mapcalc_list:
     print(mapcalc.popen.returncode)
        
GRASS_raster_files= [file for line in gs.list_strings(type='raster') for file in [line.split("@")[0]]]
GRASS_vector_files = [file for line in gs.list_strings(type='vector') for file in [line.split("@")[0]]]
```
