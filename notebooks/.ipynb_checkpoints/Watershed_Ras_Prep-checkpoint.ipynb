## Export RAS information per HUC12

```python
#initialize output names
basin_name = Project_Area+'_domains_b.shp'
stream_name = Project_Area+'stream_b.shp'

#convert basins to polygon # attach HUC12 code for modeling
gs.run_command('r.to.vect',input = basins_b,output = v_basins_b, type = 'area')
gs.run_command('v.db.join',map = v_basins_b,column = 'value', other_table = pour_points_upstream, other_column = 'cat', subset_columns = ['HUC_12','HU_12_DS'])
gs.run_command('v.out.ogr', input=  v_basins_b ,output = vector_dir/basin_name,format = 'ESRI_Shapefile')

#streams up to 1 sq mile for breaklines
gs.run_command('v.db.renamecolumn',map = v_stream_b,column='stream_type,st') #shapefile can't handle
gs.run_command('v.out.ogr', input=  v_stream_b ,output = vector_dir/stream_name,type='line',format = 'ESRI_Shapefile')

#export terrain for model
dem_name = DEM+'.tif'
gs.run_command('r.out.gdal', input= DEM, output= raster_dir/dem_name)

export_vectors = False
export_rasters = False
if export_vectors is True:
    gs.run_command('v.out.ogr', input=  v_basins_b ,output = vector_dir/'domains_b.shp',format = 'ESRI_Shapefile')
    gs.run_command('v.db.renamecolumn',map = v_stream_b,column='stream_type,st')
    gs.run_command('v.out.ogr', input=  v_stream_b ,output = vector_dir/'stream_b.shp',type='line',format = 'ESRI_Shapefile')
    gs.run_command('v.db.renamecolumn',map = upstream_nhd,column='SHAPE_Length,sl')
    gs.run_command('v.out.ogr', input=  upstream_nhd ,output = vector_dir/'upstream_nhd_carve.shp',format = 'ESRI_Shapefile')
    gs.run_command('v.out.ogr', input=  pour_points_upstream_moved ,output = vector_dir/'pour_points_upstream_moved.shp',format = 'ESRI_Shapefile')
if export_rasters is True:
    gs.run_command('r.out.gdal', input= DEM_b, output= raster_dir/'dem_burned.tif')
```
